```python
import math

with open('day_17_input') as f:
    line = f.readline().strip()
    line = line[len('target area: '):]
    x, _, y = line.partition(', ')
    _, _, x_vals = x.partition('x=')
    x_min_s, _, x_max_s = x_vals.partition('..')
    x_min = int(x_min_s)
    x_max = int(x_max_s)
    _, _, y_vals = y.partition('y=')
    y_min_s, _, y_max_s = y_vals.partition('..')
    y_min = int(y_min_s)
    y_max = int(y_max_s)

print(f'x_min={x_min} x_max={x_max} y_min={y_min} y_max={y_max}')
```

```python
def projected_x_vals(initial_x_velocity):
    xpos = 0
    xvel = initial_x_velocity
    result = []
    while xvel != 0:
        xpos += xvel
        if xvel < 0:
            xvel += 1
        else:
            xvel -= 1
        result.append(xpos)
    return result, xpos

print(projected_x_vals(7))
print(projected_x_vals(-7))
```

```python
def projected_y_vals(initial_y_velocity, min_y):
    ypos = 0
    yvel = initial_y_velocity
    result = []
    while ypos >= min_y:
        ypos += yvel
        result.append(ypos)
        yvel -= 1
    return result
print(projected_y_vals(2, y_min))
```

```python
possible_indices = []
for x in range(x_max):
    x_possible_indices, end = projected_x_vals(x)
    for index, possibility in enumerate(x_possible_indices):
        if possibility <= x_max and possibility >= x_min:
            possible_indices.append((x, index + 1, possibility == end))
print(possible_indices)

def search_on_iters(iters):
    # print(f'Checking iters {iters}')
    gravity = (iters * (iters - 1)) / 2
    y_upper = (y_max + gravity) / iters
    y_lower = (y_min + gravity) / iters
    biggest_y = math.floor(y_upper)
    if biggest_y >= y_lower:
        # print(f'  OK {y_lower} {y_upper}')
        return biggest_y
    # print(f'  BAIL {y_lower} {y_upper}')
    return None

max_y = None
for possible_index in possible_indices:
    x_pos, iters, repeats = possible_index
    biggest_y = search_on_iters(iters)
    while iters < 10000:
        if biggest_y is not None and (max_y is None or biggest_y > max_y):
            max_y = biggest_y
        if repeats:
            iters += 1
            biggest_y = search_on_iters(iters)
        else:
            break
    print(f'{iters} {biggest_y}')
print(max_y)
print((max_y * (max_y + 1)) / 2)
```

```python
possible_indices = []
for x in range(x_max + 1):
    x_possible_indices, end = projected_x_vals(x)
    for index, possibility in enumerate(x_possible_indices):
        if possibility <= x_max and possibility >= x_min:
            possible_indices.append((x, index + 1))
            if possibility == end:
                while index < 1000:
                    index += 1
                    possible_indices.append((x, index + 1))
# print(possible_indices)

def search_on_iters(iters):
    # print(f'Checking iters {iters}')
    gravity = (iters * (iters - 1)) / 2
    y_upper = (y_max + gravity) / iters
    y_lower = (y_min + gravity) / iters
    biggest_y = math.floor(y_upper)
    smallest_y = math.ceil(y_lower)
    return range(smallest_y, biggest_y + 1)

count = 0
possibilities = set()
for possible_index in possible_indices:
    # print(possible_index)
    x_pos, iters = possible_index
    # print(f'A: {x_pos} {iters}')
    for possible_y in search_on_iters(iters):
        possibilities.add((x_pos, possible_y))
        # print(f'B: {x_pos}, {possible_y}')
        count += 1
print(count)
print(len(possibilities))
```
