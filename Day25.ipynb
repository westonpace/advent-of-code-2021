```python
with open('day_25_input') as f:
    lines = [line.strip() for line in f.readlines()]
    
print(lines)
```

```python
east_cuces_template = []
south_cuces_template = []
for row, line in enumerate(lines):
    for col, val in enumerate(line):
        if val == '>':
            east_cuces_template.append((row, col))
        elif val == 'v':
            south_cuces_template.append((row, col))
            
print(east_cuces_template)
print(south_cuces_template)
```

```python
grid = [[c for c in line] for line in lines]

width = len(grid[0])
height = len(grid)

east_cuces = east_cuces_template.copy()
south_cuces = south_cuces_template.copy()

def print_map(grid):
    for row in grid:
        print(''.join(row))

def get_destination(cuce, axis):
    dst = [cuce[0], cuce[1]]
    dst[axis] += 1
    if axis == 0 and dst[axis] == height:
        dst[axis] = 0
    elif axis == 1 and dst[axis] == width:
        dst[axis] = 0
    return tuple(dst)

def is_free(grid, spot):
    return grid[spot[0]][spot[1]] == '.'

def get_moveable_cuces(grid, cuces, axis):
    moveable_cuces = {}
    for cuce in cuces:
        dst = get_destination(cuce, axis)
        if is_free(grid, dst):
            moveable_cuces[cuce] = dst
    return moveable_cuces

def mark_cuce(grid, spot, axis):
    marker =  '>' if axis == 1 else 'v'
    grid[spot[0]][spot[1]] = marker
    
def clear_cuce(grid, spot):
    grid[spot[0]][spot[1]] = '.'

def move_cuces(grid, cuces, axis):
    moveable_cuces = get_moveable_cuces(grid, cuces, axis)
    new_cuces = []
    for cuce in cuces:
        if cuce in moveable_cuces:
            dst = moveable_cuces[cuce]
            new_cuces.append(dst)
            clear_cuce(grid, cuce)
            mark_cuce(grid, dst, axis)
        else:
            new_cuces.append(cuce)
    return new_cuces, len(moveable_cuces)

print_map(grid)
print()

num_days = 0
while True:
    east_cuces, east_moved = move_cuces(grid, east_cuces, 1)
    south_cuces, south_moved = move_cuces(grid, south_cuces, 0)
    # print_map(grid)
    # print()
    num_days += 1
    if east_moved + south_moved == 0:
        break
        
print(f'Halted after {num_days}')
```
