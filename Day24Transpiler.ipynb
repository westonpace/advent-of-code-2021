```python
def prep_arg(arg):
    if arg[0] >= 'a' and arg[0] <= 'z':
        return f"'{arg}'"
    return arg

with open('day_24_input') as f:
    for line in f.readlines():
        parts = line.strip().split()
        args = ','.join([prep_arg(arg) for arg in parts[1:]])
        print(f'{parts[0]}({args})')
```
