```python
import itertools

with open('day_21_simple_input') as f:
    _, _, p1_str = f.readline().strip().partition(': ')
    _, _, p2_str = f.readline().strip().partition(': ')
```

```python
roll_counter = 1
times_rolled = 0

p1 = int(p1_str) - 1
p2 = int(p2_str) - 1


def roll():
    global roll_counter, times_rolled
    times_rolled += 3
    r = 0
    for _ in range(3):
        r += roll_counter
        roll_counter += 1
        if roll_counter > 100:
            roll_counter = 1
    return r

def take_turn(start_pos):
    r = roll()
    new_pos = (start_pos + r) % 10
    print(f'Rolled {r} to land at {new_pos}')
    return new_pos

p1_score = 0
p2_score = 0

while p1_score < 1000 and p2_score < 1000:
    print(f'{p1} {p2} {p1_score} {p2_score}')
    p1 = take_turn(p1)
    p1_score += (p1 + 1)
    if p1_score >= 1000:
        break
    p2 = take_turn(p2)
    p2_score += (p2 + 1)
    
print(f'END {p1_score} {p2_score} {times_rolled}')
```

```python
print(min(p1_score, p2_score) * times_rolled)
```
